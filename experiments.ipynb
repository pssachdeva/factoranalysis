```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
from sklearn.decomposition import FactorAnalysis

from fagm import FactorAnalysisGM
```

```python
fagm = FactorAnalysisGM(N=6, K=3, rho_b=0.1, rho_c=0.3)
```

```python
N = 20
rho_b = 0.1
rho_c = 0.3
reps = 50

Ks = np.arange(1, 6)
n_Ks = Ks.size

n_components = np.arange(1, 11)
comp_size = n_components.size

vanilla_train = np.zeros((n_Ks, reps, comp_size))
vanilla_test = np.zeros((n_Ks, reps, comp_size))
iter_train = np.zeros((n_Ks, reps, comp_size))
iter_test = np.zeros((n_Ks, reps, comp_size))
```

```python
for K_idx, K in enumerate(Ks):
    print(K_idx)
    fagm = FactorAnalysisGM(N=N, K=K, rho_b=rho_b, rho_c=rho_c)
    for rep in range(reps):
        
        components = np.zeros((comp_size, 20))
        psi = np.zeros(20)
        n_samples = 500
        # generate training set
        train = fagm.sample(n_samples=200)
        # generate test test
        test = fagm.sample(n_samples=1000)
        # copy noise
        train_copy = np.copy(train)
        var = np.var(train, axis=0)

        for idx, cur_component in enumerate(n_components):
            # apply num of components
            fa = FactorAnalysis(
                n_components=cur_component,
                tol=1e-4,
                max_iter=5000
            )
            fa.fit(train)
            # train and test
            vanilla_train[K_idx, rep, idx] = fagm.ll(train=train, test=train, L=fa.components_, psi=fa.noise_variance_)
            vanilla_test[K_idx, rep, idx] = fagm.ll(train=train, test=test, L=fa.components_, psi=fa.noise_variance_)

            # iterate
            fa_iter = FactorAnalysis(
                n_components=1,
                tol=1e-4,
                max_iter=5000
            )
            fa_iter.fit(train_copy)

            # estimate latent space
            shared_v_hat = np.dot(fa_iter.transform(train_copy), fa_iter.components_)
            # subtract shared variability
            train_copy = train_copy - shared_v_hat.reshape(train_copy.shape)

            components[idx, :] = fa_iter.components_
            psi = var - np.diag(np.dot(components.T, components))

            iter_train[K_idx, rep, idx] = fagm.ll(train=train, test=train, L=components, psi=psi)
            iter_test[K_idx, rep, idx] = fagm.ll(train=train, test=test, L=components, psi=psi)
```

```python
fig, axes = plt.subplots(5, 2, figsize=(20, 50))

vanilla_xs = np.arange(1, 11) - 0.1
iter_xs = np.arange(1, 11) + 0.1

# training set performance
for idx, K in enumerate(Ks):
    ax = axes[idx, 0]
    ax.errorbar(
        x=vanilla_xs, 
        y=np.median(vanilla_train[idx, ...], axis=0),
        yerr=np.vstack((
            np.median(vanilla_train[idx, ...], axis=0) - np.min(vanilla_train[idx, ...], axis=0), 
            np.max(vanilla_train[idx, ...], axis=0) - np.median(vanilla_train[idx, ...], axis=0)
        )),
        color='gray',
        marker='o'
    )
    ax.errorbar(
        x=iter_xs, 
        y=np.median(iter_train[idx, ...], axis=0),
        yerr=np.vstack((
            np.median(iter_train[idx, ...], axis=0) - np.min(iter_train[idx, ...], axis=0), 
            np.max(iter_train[idx, ...], axis=0) - np.median(iter_train[idx, ...], axis=0)
        )),        
        color='r',
        marker='o'
    )
    ax.set_xticks(np.arange(1, 11))
    ax.tick_params(labelsize=30)
    ax.set_ylabel(r'$K=%s$' %K, fontsize=35)

for idx, K in enumerate(Ks):
    ax = axes[idx, 1]
    ax.errorbar(
        x=vanilla_xs, 
        y=np.median(vanilla_test[idx, ...], axis=0),
        yerr=np.vstack((
            np.median(vanilla_test[idx, ...], axis=0) - np.min(vanilla_test[idx, ...], axis=0), 
            np.max(vanilla_test[idx, ...], axis=0) - np.median(vanilla_test[idx, ...], axis=0)
        )),
        color='gray',
        marker='o'
    )
    ax.errorbar(
        x=iter_xs, 
        y=np.median(iter_test[idx, ...], axis=0),
        yerr=np.vstack((
            np.median(iter_test[idx, ...], axis=0) - np.min(iter_test[idx, ...], axis=0), 
            np.max(iter_test[idx, ...], axis=0) - np.median(iter_test[idx, ...], axis=0)
        )),        
        color='r',
        marker='o'
    )
    ax.set_xticks(np.arange(1, 11))
    ax.tick_params(labelsize=30)

axes[0, 0].set_title(r'\textbf{Training Set}', fontsize=40)
axes[0, 1].set_title(r'\textbf{Test Set}', fontsize=40)
axes[4, 0].set_xlabel(r'\textbf{N. Iterations/N. Latent Factors}', fontsize=30)
axes[4, 1].set_xlabel(r'\textbf{N. Iterations/N. Latent Factors}', fontsize=30)
fig.suptitle(r'\textbf{Iterated Factor Analysis (Red)/Ordinary Factor Analysis (Gray)}', fontsize=33)
plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.savefig('iter_vs_vanilla_fa.pdf')
```
